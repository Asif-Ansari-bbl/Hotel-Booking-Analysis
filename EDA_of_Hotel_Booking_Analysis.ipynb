```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualizing the missing values
def Visulization_of_null(data):

  plt.figure(figsize=(14,5))
  sns.heatmap(data.isnull(), cbar = True,yticklabels=False)
  plt.title('Missing Values in columns \n', size=20, weight = 'bold' )
  plt.xlabel('Name of COlumns', size=20, weight='bold')
  plt.show()

Visulization_of_null(hotel_df)
```

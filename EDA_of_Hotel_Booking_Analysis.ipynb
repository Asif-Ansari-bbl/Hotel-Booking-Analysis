### Variables Description 

####**Description of all the variable persent in the dataset**

*  **lead_time:** Number of days before the actual arrival of the guests

*  **arrival_date_month:** Month of arrival date

*  **arrival_date_week_number:** Week number of year for arrival date

*  **arrival_date_day_of_month:** Day of arrival date

*  **stays_in_weekend_nights:** Number of weekend nights(Saturday or Sunday) spent at the hotel by the guests.

*  **stays_in_weel_nights:** Number of weeknights(Monday to Friday) spent at the hotel by the guests.

*  **adults:** Number of adults among the guests

*  **children:** Number of children

*  **babies:** Number of babies

*  **meal:** Type of meal booked

*  **country:** country of the guests

*  **market_segment:** Designation of market segment

*  **distribution_channel:** Name of booking distribution channel

*  **is_repeated_guest:** If the booking was from a repeated guest(1) or not(0)

*  **previous_cancellation:** Number of previous bookings that were cancelled by the customer prior to the current booking

*  **previous_bookings_not_cancelled:** Number of previous bookins not cancelled by the customer prior to the current bookin

*  **reserved_room_type:** Code from room type reserved

*  **assigned_room_type:** Code of room type assigned

*  **booking_changes:** Number of changes made to the booking

*  **deposit_type:** Type of deposite made by the guest

* **agent:** ID of travel agent who made the booking

*  **comapny:** ID of the company that made the booking

*  **days_in_waiting_list:** Number of the days the booking was in the waiting list

*  **customer_type:** Type of customer, assuming one of four categories

*  **adr:** Average daily rate

*  **required_car_parking_spaces:** Number of car parking spaces required for the customer

*  **total_of_special_requesrs:** Number of special requests made by the customer

*  **reservation_statuse:** Reservation status(Canceled, check-out or no-show)

*  **reservation_status_date:** Date at which the last reservation status was updated

## 3. ***Data Wrangling***

### Data Wrangling Code

```python
hotel_df.columns
```

```python
# Write your code to make your dataset analysis ready.
Null_value_column=['company','agent','children','country'] 
```

```python
# Haandle Null Values 

def Percentage_of_null_values(dataset, Null_column):


  result= dataset[Null_column].isna().sum()/len(dataset)*100
  print(result,'%')

Percentage_of_null_values(hotel_df, Null_value_column)
```

```python
# Drop company column because 93% of the row are having null values
hotel_df= hotel_df.drop(columns='company')

# Handle Null value of Agent with 0
hotel_df['agent'].fillna(0, inplace=True)
hotel_df['agent'].astype(int)

# Handle null value of Children with mean
hotel_df['children'].fillna(hotel_df.children.mean(), inplace=True)
hotel_df['children'] = hotel_df['children'].astype(int)

# Handle null value of Coutry  with Orther
hotel_df['country'].fillna('Orther', inplace= True)
```

```python
hotel_df.isnull().sum().sum()
```

```python
hotel_df.columns
```

```python
# Adding total staying days in hotels
hotel_df['total_stay'] = hotel_df['stays_in_weekend_nights']+hotel_df['stays_in_week_nights']

# Adding total people num as a column
hotel_df['total_people'] = hotel_df['adults']+ hotel_df['children']+hotel_df['babies']
```

```python
hotel_df.columns
```

```python
#Type Of hotel
Type_of_hotel =hotel_df['hotel'].value_counts()
print('Type of Hotel')
print(Type_of_hotel, '\n ','--'*100)

# Is Cancelled
print('\nCanellitation')
Is_canelled = hotel_df['is_canceled'].value_counts()
print(Is_canelled, '\n')

# Repeated Guest
print('\nRepeatetion')
Is_repeated_guest = hotel_df['is_repeated_guest'].value_counts()
print(Is_repeated_guest, '\n')


#Yearly Distribution of Data
print('\nYearly Disribution')
yearly_count = hotel_df['arrival_date_year'].value_counts()
print(yearly_count, '\n')

# Deposite Type
print('\nDeposite Type')
Deposite_type = hotel_df['deposit_type'].value_counts()
print(Deposite_type, '\n')
```

```python
bookings_by_months=hotel_df.groupby(['arrival_date_month'])['hotel'].count().reset_index().rename(columns={'hotel':"Counts of booking"})
bookings_by_months
```

```python
bookings_by_months=hotel_df.groupby(['arrival_date_month'])['hotel'].count().reset_index().rename(columns={'hotel':"Counts of booking"})
sequence_of_months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
bookings_by_months['arrival_date_month']=pd.Categorical(bookings_by_months['arrival_date_month'],categories=sequence_of_months,ordered=True)
bookings_by_months=bookings_by_months.sort_values('arrival_date_month')
bookings_by_months
```

```python
City_df =pd.DataFrame(hotel_df[hotel_df['hotel'] =='City Hotel'])
Resort_df =pd.DataFrame(hotel_df[hotel_df['hotel'] =='Resort Hotel'])
```

```python
resort_guest = Resort_df['arrival_date_month'].value_counts().reset_index() 
resort_guest.columns=['month','no of guests']
resort_guest

city_guest = City_df['arrival_date_month'].value_counts().reset_index()
city_guest.columns=['month','no of guests']
city_guest

final_guest=resort_guest.merge(city_guest, on = 'month')                              #merge resort guest and city guest month wise
final_guest.columns=['month','no of guests in resort','no of guest in city hotel']
final_guest
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
final_guest['month'] = pd.Categorical(final_guest['month'], categories=months, ordered=True)
```
